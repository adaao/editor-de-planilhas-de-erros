```python
import pandas as pd
import os
import time
import string
```

```python
def filtrarPlanilhas(diretorio):
    if os.path.exists(diretorio):
        arquivos = [os.path.join(diretorio, arquivo) for arquivo in os.listdir(diretorio)]
        planilhas = [arq for arq in arquivos if os.path.isfile(arq) and arq.lower().endswith('.xlsx')]
        return planilhas
    else:
        print('Os diretorios ou os arquivos xlsx não foram encontrados...')
        input('Pressione enter para sair')
        exit()
```

```python
def formata_celulas():  
  cell_format = workbook.add_format({
    'align':'center',
    'valign':'vcenter',
  })

  cell_format.set_font_name('Poppins')
  cell_format.set_font_size(14)
  cell_format.set_text_wrap()

  for col_num, values in enumerate(df.columns.values):
    worksheet.set_column(get_entire_col(col_num), None, cell_format)
```

```python
def get_entire_col(col_num):
  alfabeto_maiusculo = list(string.ascii_uppercase)
  return alfabeto_maiusculo[col_num] + ':' + alfabeto_maiusculo[col_num]
```

```python
def formata_titulos():
  header_format = workbook.add_format({
    'font_size':14,
    'bold': True,
    'font_color':'white',
    'bg_color':'#E8418E',
  })

  header_format.set_align('center')
  header_format.set_align('vcenter')
  header_format.set_font_name('Poppins')
  header_format.set_text_wrap()
  
  for col_num, value in enumerate(df.columns.values):
    worksheet.write(0, col_num, value, header_format)
```

```python
def ajusta_largura_colunas():
  for col_num, col in enumerate(df.columns.values):
    # encontra a largura da maior célula da coluna
    column_len = df[col].astype(str).str.len().max()
    
    # Seta a largura se o cabeçalho for maior
    # do que o tamanho máximo dos valores
    pixels_por_letra = 10
    margem_em_pixels = 20
    tamanho_maximo_em_pixels = 600
    if len(col) > column_len:
      column_len = (len(col) * pixels_por_letra) + margem_em_pixels
    elif column_len > 60:
      column_len = tamanho_maximo_em_pixels
    else:
      column_len = column_len * pixels_por_letra

    worksheet.set_column_pixels(col_num, col_num, column_len)
    
```

```python
def renomea_arquivos():
  n_clientes = [('fslf', 'FSLF'),
                ('pos.tuiuti', 'Pós UTP'),
                ('zu.ac.ae', 'Zayed'),
                ('redballoon', 'Red Balloon'),
                ('blogdoenem', 'Blog do ENEM'),
                ('tuiuti', 'Tuiuti'),
                ('unifacisa', 'Unifacisa'),
                ('unit', 'UNIT'),
                ('mkt4edu', 'Mkt4edu'),
                ('mdx', 'MDX')
               ]
  
  nome_dos_erros = [('internal_broken_links', 'Links internos quebrados', 'Internal broken links'),
                    ('duplicate_meta_description', 'Meta descrição duplicada', 'Duplicate meta description'),
                    ('duplicate_title_tag', 'Tag de título duplicada', 'Duplicate title tag'),
                    ('external_broken_links', 'Links externos quebrados', 'External broken links'),
                    ('302_307_redirect', 'Redirecionamentos temporários', '302 307 redirects'),
                    ('external_images_are_broken', 'Imagens externas quebradas', 'External broken images'),
                    ('h1_and_title_tags_have_duplicate', 'Tags h1 e de títulos estão iguais', 'h1 and title tag have duplicate content'),
                    ('http_4xx_client_errors', 'Páginas que retornam erro 4xx', 'http 4xx client errors'),
                    ('images_without_an_alt_attribute', 'Imagens sem atributo alt', 'Images without an alt atribute'),
                    ('internal_images_are_broken', 'Imagens internas quebradas', 'Internal images are broken'),
                    ('links_to_http_pages', 'Links para páginas HTTP', 'Links to http pages'),
                    ('missing_h1_heading', 'Páginas sem tag h1', 'Missing h1 heading'),
                    ('mixed_content', 'Páginas com conteúdo misto', 'Mixed content'),
                    ('text_html_ratio', 'Páginas com baixa proporção texto-html', 'Text html ratio'),
                    ('meta_description_tag_is_missing', 'Páginas sem meta descrição', 'Meta description tag is missing'),
                    ('title_element_is_too_long', 'Páginas com título grando demais', 'Title element is too long'),
                    ('title_element_is_too_short', 'Páginas com título muito curto', 'Title element is too short'),
                    ('underscores_in_urls', 'Páginas com underscore na URL', 'Underscors in urls'),
                    ('unminified_javascript_and_css_files', 'Arquivos javascrit e css não minificados', 'Unminified javascript and css files'),
                    ('too_long_url', 'URL muito longa', 'Too long url'),
                    ('duplicate_content', 'Conteúdo duplicado', 'Duplicate content'),
                    ('redirect_chains_and_loops', 'Cadeias de loops e redirecionamentos', 'Redirect chains and loops'),
                    ('title_tag_is_missing_or_empty', 'Tag de título faltando ou vazia', 'Title tag is missing or empty'),
                    ('language_is_not_specified', 'Linguagem não especifidada', 'Language is not specified')
                   ]
  
  todos_arquivos = os.listdir(diretorio)
  
  for arq in todos_arquivos:
    if arq.lower().endswith('.xlsx'):
      arquivo_renomeado = ''
      
      for n_cli in n_clientes:
        
        #verificar se existe a substring com o nome do cliente
        #no nome do arquivo
        if n_cli[0] in arq: 
          arquivo_renomeado = n_cli[1] + ' - '
          
          for n_err in nome_dos_erros:
            
            #verificar se existe a substring com o nome do erro
            #no nome do arquivo
            if n_err[0] in arq:
              
              if n_cli[1] == 'Zayed' or n_cli[1] == 'MDX':
                arquivo_renomeado = arquivo_renomeado + n_err[2] + '.xlsx'
              else:
                arquivo_renomeado = arquivo_renomeado + n_err[1] + '.xlsx'
              
              arquivo_renomeado = os.path.join(diretorio, arquivo_renomeado)
              arquivo_original = os.path.join(diretorio, arq)
              try:
                os.rename(arquivo_original, arquivo_renomeado)
              except:
                pass
              finally:
                pass
```
