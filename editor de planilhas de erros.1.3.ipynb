```python
import pandas as pd
import os
import time
import string
```

```python
def filtrarPlanilhas(diretorio):
    if os.path.exists(diretorio):
        arquivos = [os.path.join(diretorio, arquivo) for arquivo in os.listdir(diretorio)]
        planilhas = [arq for arq in arquivos if os.path.isfile(arq) and arq.lower().endswith('.xlsx')]
        return planilhas
    else:
        print('Os diretorios ou os arquivos xlsx não foram encontrados...')
        input('Pressione enter para sair')
        exit()
```

```python
def formata_celulas():  
  cell_format = workbook.add_format({
    'align':'center',
    'valign':'vcenter',
  })

  cell_format.set_font_name('Poppins')
  cell_format.set_font_size(14)
  cell_format.set_text_wrap()

  for col_num, values in enumerate(df.columns.values):
    worksheet.set_column(get_entire_col(col_num), None, cell_format)
```

```python
def get_entire_col(col_num):
  alfabeto_maiusculo = list(string.ascii_uppercase)
  return alfabeto_maiusculo[col_num] + ':' + alfabeto_maiusculo[col_num]
```

```python
def formata_titulos():
  header_format = workbook.add_format({
    'font_size':14,
    'bold': True,
    'font_color':'white',
    'bg_color':'#E8418E',
  })

  header_format.set_align('center')
  header_format.set_align('vcenter')
  header_format.set_font_name('Poppins')
  header_format.set_text_wrap()
  
  for col_num, value in enumerate(df.columns.values):
    worksheet.write(0, col_num, value, header_format)
```

```python
def ajusta_largura_colunas():
  for col_num, col in enumerate(df.columns.values):
    # encontra a largura da maior célula da coluna
    column_len = df[col].astype(str).str.len().max()
    
    # Seta a largura se o cabeçalho for maior
    # do que o tamanho máximo dos valores
    pixels_por_letra = 10
    margem_em_pixels = 20
    tamanho_maximo_em_pixels = 600
    if len(col) > column_len:
      column_len = (len(col) * pixels_por_letra) + margem_em_pixels
    elif column_len > 60:
      column_len = tamanho_maximo_em_pixels
    else:
      column_len = column_len * pixels_por_letra

    worksheet.set_column_pixels(col_num, col_num, column_len)
    
```

```python
def renomea_arquivos():
  n_clientes = [('fslf', 'FSLF'),
                ('pos.tuiuti', 'Pós UTP'),
                ('zu.ac.ae', 'Zayed'),
                ('redballoon', 'Red Balloon'),
                ('blogdoenem', 'Blog do ENEM'),
                ('tuiuti', 'Tuiuti'),
                ('unifacisa', 'Unifacisa'),
                ('unit', 'UNIT'),
                ('mkt4edu', 'Mkt4edu'),
                ('mdx', 'MDX')
               ]
  
  nome_dos_erros = [('internal_broken_links', 'Links internos quebrados'),
                    ('duplicate_meta_description', 'Meta descrição duplicada'),
                    ('duplicate_title_tag', 'Tag de título duplicada'),
                    ('external_broken_links', 'Links externos quebrados'),
                    ('302_307_redirect', 'Redirecionamentos temporários'),
                    ('external_images_are_broken', 'Imagens externas quebradas'),
                    ('h1_and_title_tags_have_duplicate', 'Tags h1 e de títulos estão iguais'),
                    ('http_4xx_client_errors', 'Páginas que retornam erro 4xx'),
                    ('images_without_an_alt_attribute', 'Imagens sem atributo alt'),
                    ('internal_images_are_broken', 'Imagens internas quebradas'),
                    ('links_to_http_pages', 'Links para páginas HTTP'),
                    ('missing_h1_heading', 'Páginas sem tag h1'),
                    ('mixed_content', 'Páginas com conteúdo misto'),
                    ('text_html_ratio', 'Páginas com baixa proporção texto-html'),
                    ('meta_description_tag_is_missing', 'Páginas sem meta descrição'),
                    ('title_element_is_too_long', 'Páginas com título grando demais'),
                    ('title_element_is_too_short', 'Páginas com título muito curto'),
                    ('underscores_in_urls', 'Páginas com underscore na URL'),
                    ('unminified_javascript_and_css_files', 'Arquivos javascrit e css não minificados'),
                    ('too_long_url', 'URL muito longa'),
                    ('duplicate_content', 'Conteúdo duplicado'),
                    ('redirect_chains_and_loops', 'Cadeias de loops e redirecionamentos'),
                    ('title_tag_is_missing_or_empty', 'Tag de título faltando ou vazia')
                   ]
  
  todos_arquivos = os.listdir(diretorio)
  
  for arq in todos_arquivos:
    if arq.lower().endswith('.xlsx'):
      arquivo_renomeado = ''
      
      for n_cli in n_clientes:
        
        #verificar se existe a substring com o nome do cliente
        #no nome do arquivo
        if n_cli[0] in arq: 
          arquivo_renomeado = n_cli[1] + ' - '
          
          for n_err in nome_dos_erros:
            
            #verificar se existe a substring com o nome do erro
            #no nome do arquivo
            if n_err[0] in arq:
              arquivo_renomeado = arquivo_renomeado + n_err[1] + '.xlsx'
              arquivo_renomeado = os.path.join(diretorio, arquivo_renomeado)
              arquivo_original = os.path.join(diretorio, arq)
              os.rename(arquivo_original, arquivo_renomeado)

renomea_arquivos()
```

```python
renomea_arquivos()
planilhas = filtrarPlanilhas(diretorio)
colunas = ['Discovered']

for planilha in planilhas:
  df = pd.read_excel(planilha)
  time.sleep(1)

  #remove_colunas()
  try:
    df = df.drop(columns = colunas)
    time.sleep(1)
  except:
    print('As colunas {} do arquivo {} não foram encontradas'.format(colunas, planilha))
  finally:
    pass
  
  writer = pd.ExcelWriter(planilha, engine='xlsxwriter')
  df.to_excel(writer, sheet_name='Sheet1', index=False, header=True)
  workbook = writer.book
  worksheet = writer.sheets['Sheet1']

  renomeia_colunas()

  formata_celulas()

  formata_titulos()
  
  ajusta_largura_colunas()
 
  print(planilha)
  writer.close()
```
